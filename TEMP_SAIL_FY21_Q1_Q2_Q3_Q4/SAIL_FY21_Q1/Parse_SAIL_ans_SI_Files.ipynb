```python
# import necessary libraries
import pandas as pd
import os
import glob
```

```python
# use glob to get all the csv files
# in the folder
path = os.getcwd()
csv_files = glob.glob(os.path.join(path, "SAIL/*.xlsx"))
  
# loop over the list of csv files
for f in csv_files:
    
    # read the csv file
    df_SAIL = pd.read_excel(f, header=8)
    df_SAIL.drop([0,1,2,3,4,5,6,7], axis=0, inplace=True)
    df_SAIL.drop("Unnamed: 0", axis=1, inplace=True)
    df_SAIL.drop("Unnamed: 2", axis=1, inplace=True)
    df_SAIL.drop("Unnamed: 3", axis=1, inplace=True)
    df_SAIL.drop("10th | 50th | 90th ptile", axis=1, inplace=True)
    
    hosp_city = df_SAIL.columns[3]
    print('City:',hosp_city)

    # print the location and filename
    print('Location:', f)
    print('File Name:', f.split("\\")[-1])
      
    # print the content
    print('Content:')
    display(df_SAIL.head(50))
    print()
```

```python
# use glob to get all the csv files
# in the folder
path = os.getcwd()
csv_files = glob.glob(os.path.join(path, "SI/*.xlsx"))
  
# loop over the list of csv files
for f in csv_files:
    
    # read the csv file
    df_SI = pd.read_excel(f, header=8)
    df_SI.drop([0,1,2,3,4,5,6,7], axis=0, inplace=True)
    df_SI.drop("Unnamed: 0", axis=1, inplace=True)
    df_SI.drop("Unnamed: 2", axis=1, inplace=True)
    df_SI.drop("Unnamed: 3", axis=1, inplace=True)
    df_SI.drop("10th | 50th | 90th ptile", axis=1, inplace=True)
    
    # print the location and filename
    print('Location:', f)
    print('File Name:', f.split("\\")[-1])
      
    # print the content
    print('Content:')
    display(df_SI.head(50))
    print()
```

```python
df_SAIL_SI = pd.concat([df_SAIL, df_SI])
df_SAIL_SI.head(20)
```
