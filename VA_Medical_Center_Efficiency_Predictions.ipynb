```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv(Path('../provisional_database.csv'))
df.head()
```

```python
# First, let's visually inspect the relationship between Expenditure and Efficiency:
plt.scatter(df.Expenditures, df.Efficiency)
plt.xlabel('Expenditures')
plt.ylabel('Efficiency')
plt.show()
```

```python
# plot the relationship between patient experience and Efficiency:
plt.scatter(df.patient_experience, df.Efficiency)
plt.xlabel('patient experience')
plt.ylabel('Efficiency')
plt.show()
```

```python
# plot the relationship between employee satisfaction and Efficiency:
plt.scatter(df.employee_satisfaction, df.Efficiency)
plt.xlabel('employee satisfaction')
plt.ylabel('Efficiency')
plt.show()
```

```python
# Next, we assign the target variable, or the Efficiency column, to y.
y = df.Efficiency

# assign the rest of the columns to X
X = df.drop(columns= "Efficiency")
```

```python
# create an instance of a linear regression model
model = LinearRegression()
```

```python
# fit or train the data
model.fit(X, y)
```

```python
# generate predictions 
y_pred = model.predict(X)
print(y_pred.shape)
```

```python
# plot the predictions as a red line over the data points
plt.scatter(X, y)
plt.plot(X, y_pred, color='red')
plt.show()
```

```python
# slope of the line is "model.coef_"
# y-intercept of the line is "model.intercept_"
print(model.coef_)
print(model.intercept_)
```
